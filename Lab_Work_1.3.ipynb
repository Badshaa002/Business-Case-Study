## Q1: Normal Distribution, Histogram, KDE & Bell Curve

```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate normally distributed data
np.random.seed(42)
data = np.random.normal(50, 10, 1000)

# Mean and standard deviation
mean = np.mean(data)
std = np.std(data)

# Plot histogram and KDE
plt.figure(figsize=(8,5))
sns.histplot(data, bins=30, kde=True)

# Mean and standard deviation lines
plt.axvline(mean, color='red', linestyle='--', label='Mean')
plt.axvline(mean + std, color='green', linestyle=':', label='+1 Std Dev')
plt.axvline(mean - std, color='green', linestyle=':', label='-1 Std Dev')

plt.title("Normal Distribution with Histogram and KDE")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

## Q2: Percentiles & Quartiles

```python

import pandas as pd

df = sns.load_dataset('tips')
values = df['total_bill']

df

p25 = np.percentile(values, 25)
p50 = np.percentile(values, 50)
p75 = np.percentile(values, 75)

print(p25, p50, p75)

plt.figure(figsize=(8,5))
sns.histplot(values, bins=20)
plt.axvline(p25, color='green', linestyle='--')
plt.axvline(p50, color='black', linestyle='--')
plt.axvline(p75, color='red', linestyle='--')
plt.show()
```

## Q3: Five Number Summary & Boxplot

```python

minimum = values.min()
q1 = values.quantile(0.25)
median = values.median()
q3 = values.quantile(0.75)
maximum = values.max()

print(minimum, q1, median, q3, maximum)

sns.boxplot(x=values)
plt.show()

IQR = q3 - q1
outliers = values[(values < q1 - 1.5*IQR) | (values > q3 + 1.5*IQR)]
print(outliers)
```


## Q4: Inferential Statistics

Inferential statistics uses sample data to make conclusions about a population.

Examples:
- Polling
- Clinical trials
- A/B testing


```python

population = np.random.normal(100, 15, 10000)
population_mean = np.mean(population)

sample = np.random.choice(population, 100)
sample_mean = np.mean(sample)

print(population_mean, sample_mean)

sample_means = []
for i in range(1000):
    s = np.random.choice(population, 50)
    sample_means.append(np.mean(s))

plt.figure(figsize=(8,5))
sns.histplot(sample_means, kde=True)
plt.axvline(population_mean, color='red', linestyle='-')
plt.show()
```
