### What is Statistics?

Statistics is the science of collecting, organizing, analyzing, interpreting, and presenting data.
It helps convert raw data into meaningful insights for decision-making under uncertainty.


1. Predicting sales and revenue trends
2. Customer segmentation and behavior analysis
3. Fraud detection in banking and finance
4. Medical data analysis and disease prediction
5. Quality control in manufacturing
6. Recommendation systems (Netflix, Amazon)


### Structured vs Unstructured Data

| Feature     | Structured Data | Unstructured Data  |
| ----------- | --------------- | ------------------ |
| Format      | Fixed (tables)  | No fixed format    |
| Schema      | Predefined      | Not defined        |
| Storage     | RDBMS           | Data lakes / NoSQL |
| Querying    | Easy (SQL)      | Difficult          |
| Flexibility | Low             | High               |
| Examples    | Databases       | Images, emails     |


### Applications of Statistics

**1. Business & Economics**: Used for demand forecasting, profit analysis, and market research.

**2. Medicine**: Helps in clinical trials, testing new drugs, and analyzing disease patterns.

**3. Education**: Used for result analysis, grading, and evaluating teaching methods.

**4. Government**: Applied in census, budgeting, population studies, and policy planning.

**5. Engineering**: Used in quality control, reliability testing, and process improvement.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

```

```python
data = {
    "Name": ["Yash", "Disha", "Krish", "Meet"],
    "Age": [25, 30, 28, 35],
    "Department": ["HR", "IT", "Sales", "IT"],
    "Salary": [10000, 65000, 5000, 75000]
}

df = pd.DataFrame(data)
df
```

```python
df.dtypes
```

```python
numerical_data = df.select_dtypes(include="number")
categorical_data = df.select_dtypes(include="object")

numerical_data, categorical_data
```

```python
plt.hist(df["Age"], color='blue', alpha=0.7)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
df["Department"].value_counts().plot(kind="bar", color='green', alpha=0.7)
plt.title("Department Count")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

```python
df["Department"].value_counts().plot(kind="pie", autopct="%1.1f%%", startangle=140, colors=sns.color_palette("colorblind"), wedgeprops={'alpha':0.7})
plt.title("Department Percentage")
plt.ylabel("")
plt.show()
```

```python
plt.boxplot(df["Salary"], vert=True, patch_artist=True, boxprops=dict(facecolor='red', alpha=0.7))
plt.title("Salary Distribution")
plt.ylabel("Salary")
plt.show()
```

```python
population = np.random.randint(40, 100, 1000)
sample = np.random.choice(population, 100)
```

```python
population_mean = np.mean(population)
population_std = np.std(population)

sample_mean = np.mean(sample)
sample_std = np.std(sample)

population_mean, population_std, sample_mean, sample_std
```

### Population vs Sample Statistics

- **Population statistics** describe the entire dataset.
- **Sample statistics** estimate population values using a subset.
- Sample values may slightly differ due to randomness.
- Larger samples give more accurate population estimates.
